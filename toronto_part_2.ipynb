### Toronto, Canada Segmenting and Clustering - Part 2

```python
import pandas as pd
import numpy as np
```

#### Load Toronto Neighborhood data

```python
file = r'tor.csv'
df = pd.read_csv(file)
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

#### Clean table of Toronto Neighborhood

```python
df
```

#### Read Geospatial data

```python
gfile = r'Geospatial_Coordinates.csv'
df_g = pd.read_csv(gfile)
df_g
```

```python
df_g.set_index('Postal Code',inplace=True)
```

```python
df_g
```

#### Locate the Latitude and Longitude of each Postal Code and combine the dataframe

```python
lat = []
long = []

for post in df['PostalCode']:
    lat.append(df_g.loc[post][0])
    long.append(df_g.loc[post][1])
```

```python
dft = df
```

```python
dft['Latitude'] = lat
dft['Longitude'] = long
```

#### Here is the final table

```python
dft
```

#### Let's export this one too

```python
dft.to_csv('tor_lat_long.csv')
```
